```python

import numpy as np
data1 = data1.replace('?',np.NaN)
```

```python
data1.head(10)
```

```python
data1 = data1.dropna(axis=0)
```

```python
data1.isnull().sum()
```

```python
data1['SalStat']= data1['SalStat'].map({' less than or equal to 50,000':0,' greater than 50,000':1})

print(data1['SalStat'])
```

```python
data1 = pd.get_dummies(data1,drop_first = True)
```

```python
data1
```

```python
data1.columns
```

```python
x = data1.drop(['SalStat'], axis='columns', inplace=False)
y = data1['SalStat']
```

```python
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score,confusion_matrix
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error
```

```python
X_train, X_test, y_train, y_test = train_test_split(x, y, test_size=0.3)
print(X_train.shape, X_test.shape, y_train.shape, y_test.shape)
```

```python
#linear Regression
lgr=LinearRegression(fit_intercept=True)

model_lin1=lgr.fit(X_train,y_train)

salaryStat_predictions = lgr.predict(X_test)

lin_mse1 = mean_squared_error(y_test, salaryStat_predictions)
lin_rmse1 = np.sqrt(lin_mse1)
print(lin_rmse1)

r2_lin_test1=model_lin1.score(X_test,y_test)
r2_lin_train1=model_lin1.score(X_train,y_train)
print(r2_lin_test1,r2_lin_train1)

residuals1=y_test-salaryStat_predictions
sns.regplot(x=salaryStat_predictions, y=residuals1, scatter=True, 
            fit_reg=False)
residuals1.describe()
```

```python
target_names = ['less than or equal to 50,000','greater than 50,000']
```

```python
# KNN
from sklearn.neighbors import KNeighborsClassifier
knn = KNeighborsClassifier(n_neighbors = 6)
knn.fit(X_train , y_train)
prediction =knn.predict(X_test) 
from sklearn.metrics import accuracy_score
accuracy = accuracy_score(y_test, prediction)
print(accuracy)
```

```python
confusionMmatrix_knn = confusion_matrix(y_test, prediction)
print(confusionMmatrix_knn)
```

```python
import matplotlib.pyplot as plt
from sklearn.metrics import confusion_matrix
from sklearn.metrics import ConfusionMatrixDisplay
from sklearn.metrics import accuracy_score , classification_report
cm = confusion_matrix(y_test, prediction.reshape(-1,1), normalize='true')
disp = ConfusionMatrixDisplay(confusion_matrix=cm, display_labels=target_names)
fig, ax = plt.subplots(figsize=(8,6))
disp.plot(ax=ax,cmap='Blues')
```

```python
#logistic Regression 
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import classification_report ,confusion_matrix ,ConfusionMatrixDisplay
lgr = LogisticRegression()
lgr.fit(X_train , y_train)

prediction = lgr.predict(X_test)
from sklearn import metrics
from sklearn.metrics import accuracy_score
accuracy = accuracy_score(y_test,prediction)
print(accuracy)

print(classification_report(y_test, prediction.reshape(-1,1), target_names=target_names))

cm = confusion_matrix(y_test, prediction.reshape(-1,1), normalize='true')
disp = ConfusionMatrixDisplay(confusion_matrix=cm, display_labels=target_names)
fig, ax = plt.subplots(figsize=(8,6))
disp.plot(ax=ax,cmap='Blues')
```

```python
#Svm
from sklearn.svm import SVC
svc = SVC()
svc.fit(X_train, y_train.values.ravel())

prediction = svc.predict(X_test)
prediction

accuracy = metrics.accuracy_score(y_test , prediction)
print(accuracy)

print(classification_report(y_test, prediction.reshape(-1,1), target_names=target_names))

cm = confusion_matrix(y_test, prediction.reshape(-1,1), normalize='true')
disp = ConfusionMatrixDisplay(confusion_matrix=cm, display_labels=target_names)
fig, ax = plt.subplots(figsize=(8,6))
disp.plot(ax=ax,cmap='Blues')
```

```python
#navbayes 
from sklearn.naive_bayes import GaussianNB
model = GaussianNB()
model.fit(X_train, y_train)

prediction = model.predict(X_test)
prediction

from sklearn.metrics import accuracy_score
accuracy = accuracy_score(y_test, prediction)
print(accuracy)

print(classification_report(y_test, prediction.reshape(-1,1), target_names=target_names))

cm = confusion_matrix(y_test, prediction.reshape(-1,1), normalize='true')
disp = ConfusionMatrixDisplay(confusion_matrix=cm, display_labels=target_names)
fig, ax = plt.subplots(figsize=(8,6))
disp.plot(ax=ax,cmap='Blues')
```
