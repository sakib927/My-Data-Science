```python
import pandas as pd
df = pd.read_csv("/content/drive/MyDrive/weather.xlsx.csv")
df
```

```python
df
```

```python
from sklearn.preprocessing import LabelEncoder
from sklearn.naive_bayes import GaussianNB
```

```python
outlook=LabelEncoder()
Temp=LabelEncoder()
Hum=LabelEncoder()
win=LabelEncoder()
```

```python
inputs=df.drop('Play',axis='columns')
target=df['Play']
target
```

```python
inputs['outlook_n']= outlook.fit_transform(inputs['OUTLOOK'])
inputs['Temp_n']= Temp.fit_transform(inputs['TEMPERATURE'])
inputs['Hum_n']= Hum.fit_transform(inputs['HUMIDITY'])
inputs['win_n']= win.fit_transform(inputs['WINDY'])
inputs
```

```python
inputs_n=inputs.drop(['OUTLOOK','TEMPERATURE','HUMIDITY','WINDY'],axis='columns')
inputs_n
```

```python
classifier = GaussianNB()
classifier.fit(inputs_n,target)
classifier.score(inputs_n,target)
```
